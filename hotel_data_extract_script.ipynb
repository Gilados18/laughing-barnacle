```python
import bs4
import pandas as pd
import re
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.parse import urlsplit, urlunsplit

import time
import requests
```

```python



#Function that open the url with selenium, and return the page source
def getPageBySel(url):
    options = webdriver.ChromeOptions()
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--remote-debugging-port=9222")
    options.add_argument("--window-size=1920x1080")
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    page = driver.page_source
    driver.quit()
    return page

#Open the csv through path and read
csv_path="hotels_list123.csv"
limited_df = pd.read_csv(csv_path)
df = limited_df.head(100)
driver = webdriver.Chrome()
#Creating Lists
locationList=[]
starsList=[]
staffList=[]
comfortList=[]
vfmList=[]
wifiList=[]
cleanList=[]
reviewsList=[]
rateList=[]
facCountList=[]
pricesList=[]
facRatList=[]

#Loop that run on all of the cells in the csv, open the links and extract data
for row_index, row in df.iterrows():
    
    hotel_name = row['Hotel']
    hotel_link = row['Link']
    hotel_price = row['Prices']
    pricesList.append(hotel_price)
    driver.get(hotel_link)
    # Find the index of the first occurrence of "html"
    x = hotel_link.find("html")

    # Remove everything after the first occurrence of "html"
    new_url = hotel_link[:x+4]  # Include the length of "html"

    while True: #Loop that make sure that the page loaded successfully 
        temp=getPageBySel(new_url)
        soup = BeautifulSoup(temp, 'html.parser')
        if len(soup("h2",{"class":"d2fee87262 pp-header__title"}))>0:
            break
        else:
            time.sleep(1.5)
       
    

  
    
    time.sleep(1.5)
    #Extract the total rate
    rate = soup.find('div', {'class': 'b5cd09854e d10a6220b4'})
    if rate:
        rate = rate.text.strip()
    else:
        rate = "N/A"
    rateList.append(rate)
    #Extract the num of reviews
    review_div = soup.find('div', class_='d8eab2cf7f c90c0a70d3 db63693c62')
    if review_div:
        reviews_text = review_div.text.strip()
        reviews = re.search(r'\d{1,3}(?:,\d{3})*', reviews_text)
        if reviews:
            reviews = reviews.group(0)
        else:
            reviews = "N/A"
    else:
        reviews = "N/A"
    reviewsList.append(reviews)

    #Extract the clean rate
    clean = soup.find('div', id=':Rji:-label')
    if clean:
        clean = clean.text.strip()
    else:
        clean = "N/A"
    cleanList.append(clean)

    #Extract the location rate
    location = soup.find('div', id=':Rmi:-label')
    if location:
        location = location.text.strip()
    else:
        location = "N/A"
    locationList.append(location)

    #Extract the wifi rate
    wifi = soup.find('div', id=':Rni:-label')
    if wifi:
        wifi = wifi.text.strip()
    else:
        wifi = "N/A"
    wifiList.append(wifi)

    #Extract the staff rate
    staff = soup.find('div', id=':Rhi:-label')
    if staff:
        staff = staff.text.strip()
    else:
        staff = "N/A"
    staffList.append(staff)

    #Extract the comfort rate
    comfort = soup.find('div', id=':Rki:-label')
    if comfort:
        comfort = comfort.text.strip()
    else:
        comfort = "N/A"
    comfortList.append(comfort)

    #Extract the vfm rate
    vfm = soup.find('div', id=':Rli:-label')
    if vfm:
        vfm = vfm.text.strip()
    else:
        vfm = "N/A"
    vfmList.append(vfm)
    #Extract the facilites rating rate
    facRat = soup.find('div', id=':Rii:-label')
    if facRat:
        facRat = facRat.text.strip()
    else:
        facRat = "N/A"
    facRatList.append(facRat)


   

        #Extract the stars of the hotel
    elements = soup.find_all(class_="b6dc9a9e69 adc357e4f1 fe621d6382")

    # Count the number of occurrences
    stars = len(elements)
    starsList.append(stars)
    

    
  
 
  
    # Ensure all lists have the same length
    if len(locationList) != row_index+1:
        locationList.append("N/A")
    if len(starsList) != row_index+1:
        starsList.append("N/A")
    if len(staffList) != row_index+1:
        staffList.append("N/A")
    if len(comfortList) != row_index+1:
        comfortList.append("N/A")
    if len(vfmList) != row_index+1:
        vfmList.append("N/A")
    if len(wifiList) != row_index+1:
        wifiList.append("N/A")
    if len(cleanList) != row_index+1:
        cleanList.append("N/A")
    if len(reviewsList) != row_index+1:
        reviewsList.append("N/A")
    if len(rateList) != row_index+1:
        rateList.append("N/A")
    if len(facCountList) != row_index+1:
        facCountList.append("N/A")
    if len(pricesList) != row_index+1:
        pricesList.append("N/A")
    if len(facRatList) != row_index+1:
        facRatList.append("N/A")
    
df = pd.DataFrame({
    'Location':locationList, 'Stars':starsList, 'Staff':staffList, 'Comfort':comfortList, 'Value for money':vfmList,'WiFi':wifiList,'Clean':cleanList,'Reviews':reviewsList,'Rating':rateList,'Facilities Amount':facCountList,'Price':pricesList,'Facilites Rating':facRatList
})
        
        #Df to csv
df.to_csv('hotelsData1231.csv', index=False)
print(df)

    
    
  
    
```
